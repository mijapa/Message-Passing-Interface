```python
import pandas as pd
import csv

df = pd.read_csv('./cmake-build-debug/data.csv',skiprows=1, header=None).T
df.columns = df.iloc[0]                                 # Set new column names
df.drop(0,inplace=True)
original_headers = list(df.columns.values)
print(original_headers)
```

```python
df.mean()
```

```python
df.std()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
std = df.std()

p = df.mean().plot(legend=False,kind="bar",rot=45,color="blue",fontsize=16,yerr=std);
p.set_title("Average pi", fontsize=18);
p.set_xlabel("Number of repetitions", fontsize=18);
p.set_ylabel("Average Pi value", fontsize=18);
p.set_ylim(2.9,3.4);
```
